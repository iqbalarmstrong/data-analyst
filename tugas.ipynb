```python
import pandas as pd
import matplotlib.pyplot as plt


data = {
    "Month": ["Januari", "Febuari", "Maret", "April"],
    "SUM of Sold": [25.98,25.42,24.18,24.41]
}
df = pd.DataFrame(data)


plt.figure(figsize=(8, 6))  
plt.plot(df["Month"], df["SUM of Sold"], marker='o') 
# plt.ylim(0, 100)  
plt.xlabel("Month")
plt.ylabel("SUM of Sold (%)")
plt.title("Monthly Sales as a Percentage")
plt.grid(True)   


plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame from the provided data
data = {
    "Merk": ["Honda", "Mitsubishi", "Nissan", "Suzuki", "Toyota"],
    "Xpander": [0, 8.10, 0, 0, 0],
    "Swift": [0, 0, 0, 8.55, 0],
    "Pajero": [0, 8.77, 0, 0, 0],
    "March": [0, 0, 7.76, 0, 0],
    "Land Cruiser": [0, 0, 0, 0, 7.09],
    "Juke": [0, 0, 9.11, 0, 0],
    "Jazz": [7.54, 0, 0, 0, 0],
    "Innova": [0, 0, 0, 0, 10.01],
    "Ertiga": [0, 0, 0, 8.66, 0],
    "CRV": [8.21, 0, 0, 0, 0],
    "Brio": [8.10, 0, 0, 0, 0],
    "Avanza": [0, 0, 0, 0, 8.10],
}

df = pd.DataFrame(data)

# Plot the data as a grouped bar graph
ax = df.plot(x="Merk", kind="bar", stacked=True, figsize=(10, 6))
plt.ylim(0, 100)  # Set the y-axis range to be 0-100%
plt.xlabel("Merk")
plt.ylabel("SUM of Sold (%)")
plt.title("Monthly Sales by Merk")
plt.grid(True)  # Optional: Add grid lines

# Show the graph
plt.show()
```
